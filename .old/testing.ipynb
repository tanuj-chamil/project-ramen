```python
from spectrum import *
import pandas as pd
```

```python
file_path = "databases\\csv\\riboflavin.csv"
data = pd.read_csv(file_path)

molecule =  Spectrum('adenine', data['L'], data['I'])
```

```python
molecule.plot()
```

```python
molecule.denoise(polyorder=2,section_size=3)
molecule.plot()
```

```python
molecule.correct_baseline(polyorder=6)
molecule.plot()
```

```python
molecule.min_max_normalize()
molecule.plot()
```
